## DTO REF 저장

```python
definitions = data['definitions']
```

```python
def make_body(body,definitions, dto,count,is_array) :
    
    count = count +1
    pros = list(definitions[dto]['properties'].keys())
    
    tabs = ''
    for i in range(0,count-1):
        tabs = tabs + '\t'
    
    if is_array == True :
        body = body +'\n'+ tabs + 'List<'+dto+'>:[{'
    else :
        body = body +'\n'+ tabs +dto + ':{'

    for pro in pros :

        property_key_list = list(definitions[dto]['properties'][pro].keys())
        if 'type' in property_key_list :
            if definitions[dto]['properties'][pro]['type'] == 'array':
                items = list(definitions[dto]['properties'][pro]['items'].keys())
                if '$ref' in items:
                    body =  make_body(body,definitions,definitions[dto]['properties'][pro]['items']['$ref'].replace('#/definitions/',''),count,True)
                else :
                    body = body +'\n' + tabs+'\t' + pro +' : List<'+ definitions[dto]['properties'][pro]['items']['type'] +'>,'
            else :
                body = body +'\n' + tabs+'\t' + pro +' : '+ definitions[dto]['properties'][pro]['type'] +','
        else :          
            body =  make_body(body,definitions,definitions[dto]['properties'][pro]['$ref'].replace('#/definitions/',''),count,False)
            
        
    if is_array == True :
        body = body[:-1] + ' \n'+tabs+'}],'
    else :
        body = body[:-1] + ' \n'+tabs+'},'
    return body

body = ''
count = 0
```

# 워드작성

```python
doc = Document()

doc.add_heading('API 문서', level=0)
error_cnt = 0
out_table = 0
for path in tqdm(path_list) :
    method_list = list(paths[path].keys())
    for method in method_list :
        tag = paths[path][method]['tags'][0]
        summary =  paths[path][method]['summary']
        operationId =  paths[path][method]['operationId']
        
        doc.add_heading(str(out_table+1)+'.'+summary, level=5)

        table = doc.add_table(rows = 6, cols = 3)
        table.cell(0, 0).merge(table.cell(1, 0))
        table.cell(2,0).merge(table.cell(2,2))
        table.cell(3,0).merge(table.cell(3,2))
        table.cell(4,0).merge(table.cell(4,2))
        table.cell(5,0).merge(table.cell(5,2))
        table.style = doc.styles['Table Grid']

        # 0. 테이블 전체 틀
        table.rows[0].cells[0].text = tag
        doc.tables[out_table].rows[0].cells[0].paragraphs[0].alignment = WD_TABLE_ALIGNMENT.CENTER
        doc.tables[out_table].rows[0].cells[0].vertical_alignment = WD_CELL_VERTICAL_ALIGNMENT.CENTER
        table.rows[0].cells[1].text = '서비스명'
        doc.tables[out_table].rows[0].cells[1].paragraphs[0].alignment = WD_TABLE_ALIGNMENT.CENTER
        doc.tables[out_table].rows[0].cells[1].vertical_alignment = WD_CELL_VERTICAL_ALIGNMENT.CENTER
        table.rows[0].cells[2].text = summary
        doc.tables[out_table].rows[0].cells[2].paragraphs[0].alignment = WD_TABLE_ALIGNMENT.CENTER
        doc.tables[out_table].rows[0].cells[2].vertical_alignment = WD_CELL_VERTICAL_ALIGNMENT.CENTER
        table.rows[1].cells[1].text = 'Operation ID 값'
        doc.tables[out_table].rows[1].cells[1].paragraphs[0].alignment = WD_TABLE_ALIGNMENT.CENTER
        doc.tables[out_table].rows[1].cells[1].vertical_alignment = WD_CELL_VERTICAL_ALIGNMENT.CENTER
        table.rows[1].cells[2].text = operationId
        doc.tables[out_table].rows[1].cells[2].paragraphs[0].alignment = WD_TABLE_ALIGNMENT.CENTER
        doc.tables[out_table].rows[1].cells[2].vertical_alignment = WD_CELL_VERTICAL_ALIGNMENT.CENTER

        # 1. API 기본 정보
        table.rows[2].cells[0].text = '\n1) API 기본 정보'
        inner_table1 = doc.tables[out_table].rows[2].cells[0].add_table(rows = 3, cols = 8)

        inner_table1.cell(0,1).merge(inner_table1.cell(0,7))
        inner_table1.cell(1,1).merge(inner_table1.cell(1,7))
        inner_table1.cell(2,0).merge(inner_table1.cell(2,7))
        inner_table1.style = doc.styles['Table Grid']

        inner_table1.rows[0].cells[0].text = 'Method'
        inner_table1.rows[0].cells[0].paragraphs[0].alignment = WD_TABLE_ALIGNMENT.CENTER
        inner_table1.rows[0].cells[0].vertical_alignment = WD_CELL_VERTICAL_ALIGNMENT.CENTER
        inner_table1.rows[1].cells[0].text = method.upper()
        inner_table1.rows[1].cells[0].paragraphs[0].alignment = WD_TABLE_ALIGNMENT.CENTER
        inner_table1.rows[1].cells[0].vertical_alignment = WD_CELL_VERTICAL_ALIGNMENT.CENTER
        
        inner_table1.rows[0].cells[1].text = 'URI'
        inner_table1.rows[0].cells[1].paragraphs[0].alignment = WD_TABLE_ALIGNMENT.CENTER
        inner_table1.rows[0].cells[1].vertical_alignment = WD_CELL_VERTICAL_ALIGNMENT.CENTER
        inner_table1.rows[1].cells[1].text = path
        inner_table1.rows[1].cells[1].vertical_alignment = WD_CELL_VERTICAL_ALIGNMENT.CENTER

        inner_table1.rows[2].cells[0].text = '[비고]\n\n'
        inner_table1.rows[2].cells[0].paragraphs[0].alignment = WD_TABLE_ALIGNMENT.LEFT
        inner_table1.rows[2].cells[0].vertical_alignment = WD_CELL_VERTICAL_ALIGNMENT.TOP

        # 2. Query Parameter
        table.rows[3].cells[0].text = '\n2) Query Parameter'
        q_size = 0
        try :
            parameters = list(paths[path][method]['parameters'])
            for parameter in parameters :
                if parameter['in'] == 'query' or parameter['in'] == 'path':
                    q_size = q_size + 1
        except : 
            pass
        inner_table2 = doc.tables[out_table].rows[3].cells[0].add_table(rows = 1+q_size, cols = 10)

        inner_table2.cell(0,0).merge(inner_table2.cell(0,1))
        inner_table2.cell(0,2).merge(inner_table2.cell(0,4))
        inner_table2.cell(0,6).merge(inner_table2.cell(0,7))
        inner_table2.cell(0,8).merge(inner_table2.cell(0,9))
        inner_table2.style = doc.styles['Table Grid']

        inner_table2.rows[0].cells[0].text = '항목'
        inner_table2.rows[0].cells[0].paragraphs[0].alignment = WD_TABLE_ALIGNMENT.CENTER
        inner_table2.rows[0].cells[0].vertical_alignment = WD_CELL_VERTICAL_ALIGNMENT.CENTER
        inner_table2.rows[0].cells[2].text = '설명'
        inner_table2.rows[0].cells[2].paragraphs[0].alignment = WD_TABLE_ALIGNMENT.CENTER
        inner_table2.rows[0].cells[2].vertical_alignment = WD_CELL_VERTICAL_ALIGNMENT.CENTER
        inner_table2.rows[0].cells[5].text = '필수'
        inner_table2.rows[0].cells[5].paragraphs[0].alignment = WD_TABLE_ALIGNMENT.CENTER
        inner_table2.rows[0].cells[5].vertical_alignment = WD_CELL_VERTICAL_ALIGNMENT.CENTER
        inner_table2.rows[0].cells[6].text = '데이터타입'
        inner_table2.rows[0].cells[6].paragraphs[0].alignment = WD_TABLE_ALIGNMENT.CENTER
        inner_table2.rows[0].cells[6].vertical_alignment = WD_CELL_VERTICAL_ALIGNMENT.CENTER
        inner_table2.rows[0].cells[8].text = '비고'
        inner_table2.rows[0].cells[8].paragraphs[0].alignment = WD_TABLE_ALIGNMENT.CENTER
        inner_table2.rows[0].cells[8].vertical_alignment = WD_CELL_VERTICAL_ALIGNMENT.CENTER
        
        if q_size != 0 :
            i = 1
            for parameter in parameters :
                if parameter['in'] == 'query' or parameter['in'] == 'path' :
                    inner_table2.cell(i,0).merge(inner_table2.cell(i,1))
                    inner_table2.cell(i,2).merge(inner_table2.cell(i,4))
                    inner_table2.cell(i,6).merge(inner_table2.cell(i,7))
                    inner_table2.cell(i,8).merge(inner_table2.cell(i,9))

                    inner_table2.rows[i].cells[0].text = parameter['name']
                    try :
                        inner_table2.rows[i].cells[2].text = str(parameter['description'])
                    except :
                        pass
                    inner_table2.rows[i].cells[5].text =  str(parameter['required'])
                    try :
                        if str(parameter['type']) == 'array' :
                            inner_table2.rows[i].cells[6].text = 'List<'+parameter['items']['type']+'>'
                        else :
                            inner_table2.rows[i].cells[6].text =  str(parameter['type'])
                    except:
                        pass
                    try:
                        inner_table2.rows[i].cells[8].text =  str(parameter['enum'])
                    except:
                        pass
                    i = i+1
                    
        
        
        # 3. Body Parameter
        table.rows[4].cells[0].text = '\n3) Body Parameter'
        b_size = 0
        try :
            parameters = list(paths[path][method]['parameters'])
            for parameter in parameters :
                if parameter['in'] == 'body':
                    b_size = b_size + 1
        except : 
            pass
        inner_table3 = doc.tables[out_table].rows[4].cells[0].add_table(rows = b_size, cols = 1)
        if b_size != 0 :
            i = 0
            body = ''
            count = 0
            for parameter in parameters :
                if parameter['in'] == 'body':
        
                    try :
                        inner_table3.rows[i].cells[0].text = make_body(body,definitions,parameter['schema']['$ref'].replace('#/definitions/',''),count,False)[:-1]
                    except :
                        try :
                            if parameter['schema']['type'] == 'array' :
                                inner_table3.rows[i].cells[0].text = make_body(body,definitions,parameter['schema']['items']['$ref'].replace('#/definitions/','').replace('Response«','').replace('»',''),count,True)[:-1]
                            else :
                                inner_table3.rows[i].cells[0].text = '{\n\t'+parameter['name']+':'+ parameter['schema']['type'] +'\n}'
                        except :
                            error_cnt= error_cnt+1
                            pass
                    i = i+1
                    
        
        # 4. Response
        table.rows[5].cells[0].text = '\n4) Response'        
        inner_table4 = doc.tables[out_table].rows[5].cells[0].add_table(rows = 1, cols = 1)
        body=''
        count=0
        if '200' in paths[path][method]['responses']:
            try:
                schema_list = list(paths[path][method]['responses']['200']['schema'].keys())
                
                if 'type' in schema_list :
                    inner_table4.rows[0].cells[0].text = make_body(body,definitions,paths[path][method]['responses']['200']['schema']['items']['$ref'].replace('#/definitions/','').replace('Response«','').replace('»',''),count,True)[:-1]
                else :
                    inner_table4.rows[0].cells[0].text = make_body(body,definitions,paths[path][method]['responses']['200']['schema']['$ref'].replace('#/definitions/','').replace('Response«','').replace('»',''),count,False)[:-1]

            except:
                inner_table4.rows[0].cells[0].text = '\'OK\''


        
    
        
        out_table = out_table+1



print(error_cnt)


doc.save(info_title +'_API DOCS.docx')
```
